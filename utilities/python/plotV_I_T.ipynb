```python
infile_name = "/home/coure/SiPMs_QA/results/SiPM_test_newHV/SiPM_rec_2.csv"
colnames = ["Vset[CH1]", "Vout[CH1]", "ON_OFF[CH1]", "Vsetpoint[CH1]", "Iout[CH1]",
            "Vref[CH1]", "T[CH1]", "IC[CH1]", "VC[CH1]", "TC[CH1]", "date", "time", "AM/PM"]
df = pd.read_csv(infile_name, skiprows=1, sep = ' ', names=colnames )
# df.drop(columns=["date", "time", "AM/PM"], inplace=True)
# df = df.astype(float)
```

```python
# df["Vout[CH1]"] = df["Vout[CH1]"].astype(float)
# plt.plot(float(df["Vout[CH1]"]))
```

```python
V = []
I = []
for i in range(len(df["Vout[CH1]"])):
#     print(df["Vout[CH1]"][i])
    V.append(float(df["Vout[CH1]"][i]))
    I.append(float(df["Iout[CH1]"][i]))
```

```python
plt.figure(figsize = (20,6))
plt.plot(V, marker = "o")
plt.grid()
plt.xlabel("Seconds after the start of the run")
plt.ylabel("Out bias [V]")
plt.tight_layout()
plt.savefig("./Plots/outVoltage.pdf")
```

```python
plt.figure(figsize = (20,6))
plt.plot(np.array(I)/1E-3, marker = "o")
plt.grid()
plt.xlabel("Seconds after the start of the run")
plt.ylabel("Out current [mA]")
plt.savefig("./Plots/outCurrent.pdf")
```
